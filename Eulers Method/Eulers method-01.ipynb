```python
import numpy as np

# set x grid and spacing
dx = 0.01
x = np.arange(0, 1 + dx, dx)
N = len(x)

# create the y grid
y = np.zeros((N))

# set the initial condition
y[0] = 2

for i in range(0, N - 1):
    # the Euler ethod
    y[i + 1] = y[i] + dx*(x[i]*y[i])
    
print(np.transpose([x, y]))
```
